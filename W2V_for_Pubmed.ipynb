```python
from gensim.models import Word2Vec
texts = []
with open("heartdisease.txt",'r') as myfile:
    for line in myfile:
        #print(line)
        #print(type(line))
        texts.append(line)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
corpus = []
for text in texts:
    tokens = word_tokenize(text)
    print(tokens)
    corpus.append(tokens)
```

```python
model = Word2Vec(corpus, min_count=1)
```

```python
print (list(model.wv.vocab))
```

```python
print(model['disease'])
```

```python
result = model.most_similar(['disease'])
print(result)
```

```python
vocab_words = list(model.wv.vocab)
```

```python
f = open("diabetes-tfidf-w2v","w")
with open("diabetes-tfidf",'r') as myfile:
    for line in myfile:
        if line.strip() in vocab_words:
            #print(line)
            similar_words = model.most_similar([line.strip()])
            for word in similar_words:
                print(line+","+word[0]+","+str(word[1]))
                f.write(line.strip()+","+word[0]+","+str(word[1])+"\n")
f.close()
```
