```python
from gensim.models import Word2Vec
texts = []
with open("heartdisease.txt",'r') as myfile:
    for line in myfile:
        #print(line)
        #print(type(line))
        texts.append(line)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
corpus = []
for text in texts:
    tokens = word_tokenize(text)
    print(tokens)
    corpus.append(tokens)
```

```python
#model = word2vec.Word2Vec(corpus, size=100, window=20, min_count=200, workers=4)
```

```python
model = Word2Vec(corpus, min_count=1)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False
def tsne_plot(model):
    "Creates and TSNE model and plots it"
    labels = []
    tokens = []

    for word in model.wv.vocab:
        if word.isalpha() and len(word)>2 and word.lower() not in stop_words:
            tokens.append(model[word])
            labels.append(word)
    
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    plot_only = 50
    new_values = tsne_model.fit_transform(tokens[:plot_only])
    print(len(tokens))

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(20, 16)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom',fontsize=15)
    plt.show()
    plt.savefig("diabetes_vector_space.png")
```

```python
tsne_plot(model)
```

```python
vocab_old = list(model.wv.vocab)
vocab = []
for word in vocab_old:
    if len(word)>2 and word.isalpha() and word not in stop_words:
        vocab.append(word)
X = model[vocab]
```

```python
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)
```

```python
import pandas as pd
df = pd.DataFrame(X_tsne, index=vocab, columns=['x', 'y'])
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.scatter(df['x'], df['y'])
```

```python
for word, pos in df.iterrows():
    ax.annotate(word, pos)
```
